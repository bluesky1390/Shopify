```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statistics as st 

df = pd.read_excel (r'/Users/zakiehnabavi/Downloads/data.xlsx')

plt.boxplot(df['order_amount'])
plt.show()

```

```python
x= df['shop_id']
y= df['order_amount']
plt.scatter(x, y)
plt.show()
```

```python
Q1 = np.percentile(df['order_amount'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df['order_amount'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1

upper = np.where(df['order_amount'] >= (Q3+1.5*IQR))
lower = np.where(df['order_amount'] <= (Q1-1.5*IQR))
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)
```
